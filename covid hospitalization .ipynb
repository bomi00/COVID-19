```python
a = df.groupby(['Date updated']).sum()
```

```python
x1 = np.array(a['Confirmed cases'])
x1
```

```python
x1.shape
```

```python
y1.shape
```

```python
from sklearn.linear_model import LinearRegression 
linear_regressor = LinearRegression()
linear_regressor.fit(x1.reshape(-1,1), y1.reshape(-1,1))
```

```python
y1_pred = linear_regressor.predict(x1.reshape(-1,1))
```

```python
plt.scatter(x1.reshape(-1,1),y1.reshape(-1,1))
plt.plot(x1.reshape(-1,1),y1_pred,color='purple')
```
